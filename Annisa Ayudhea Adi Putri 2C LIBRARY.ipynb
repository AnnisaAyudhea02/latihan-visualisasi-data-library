```python
import seaborn as sns
import matplotlib.pyplot as plt

# Data contoh
usia = [25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 30,
         35, 40, 45, 50, 55, 60, 65, 70, 75]

gaji = [5000, 6000, 7500, 8000, 9000, 9500, 
        10000, 11000, 12000, 13000, 5500, 
        6500, 7000, 8500, 9500, 12000, 
        11000, 12000, 13000, 14000]

# Histogram
plt.figure(figsize=(6, 4))
sns.histplot(usia, bins=20, kde=True, color='pink')
plt.title('Distribusi Usia')
plt.xlabel('Usia')
plt.ylabel('Frekuensi')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data contoh
hari = ['H1', 'H2', 'H3', 'HA', 'H5', 'H6', 'H7', 'H8', 'H9', 'H10']
jumlah_penjualan = [40, 35, 50, 40, 70, 45, 65, 75, 35, 75]


#grafiknya
plt.figure(figsize=(4, 4))
plt.pie(jumlah_penjualan, labels=hari, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart Penjualan per hari')
plt.axis('equal') # Agar Lingkaran berbentuk Lingkaran sempurna
plt.show()
```

Contoh visualisasi library dari MatPlotlib grafik lingkaran (pie)

```python
import matplotlib.pyplot as plt

# Data contoh
hari = ['H1', 'H2', 'H3', 'H4', 'H5', 'H6', 'H7', 'H8', 'H9', 'H10'] 
jumlah_penjualan = [40, 30, 50, 45, 60, 55, 75, 70, 35, 65]

# grafiknya
plt.figure(figsize=(10, 5))
plt.plot(hari, jumlah_penjualan, marker='*', color='red', linestyle='-')
plt.title('Line Chart Penjualan per hari')
plt.xlabel('hari')
plt.ylabel('Jumlah Penjualan')
plt.grid(True)
plt.show()
```
